```python
import math
import numpy as np

def binomial(n, k, p_array):
    comb = math.factorial(n) / (math.factorial(k) * math.factorial(n-k))
    return np.array([ comb * p ** k * (1-p) ** (n-k) for p in p_array])

def binomial_p(n, k, p):
    comb = math.factorial(n) / (math.factorial(k) * math.factorial(n-k))
    return comb * p ** k * (1-p) ** (n-k)
```

```python
import numpy as np

p_grid = np.arange(0, 10) / 10.0
prior = np.ones((10, ))
likelihood = binomial(3, 3, p_grid)
unstd_posterior = likelihood * prior
posterior = unstd_posterior / np.sum(unstd_posterior)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x=p_grid, y=posterior)
plt.plot(p_grid, posterior)
```

```python
def plot_binomial(n, k):
    # calculate posterior
    p_grid = np.arange(0, 10) / 10.0
    prior = np.ones((10, ))
    likelihood = binomial(n, k, p_grid)
    unstd_posterior = likelihood * prior
    posterior = unstd_posterior / np.sum(unstd_posterior)
    
    # plot
    plt.scatter(x=p_grid, y=posterior)
    plt.plot(p_grid, posterior)
```

```python
plot_binomial(3, 3)
plot_binomial(4, 3)
plot_binomial(7, 5)
```

```python
def plot_binomial_masked(k, n):
    # calculate posterior
    p_grid = np.arange(0, 10) / 10.0
    mask = np.array([0 if p < 0.5 else 1 for p in p_grid])
    prior = np.ones((10, )) * mask
    likelihood = binomial(n, k, p_grid)
    unstd_posterior = likelihood * prior
    posterior = unstd_posterior / np.sum(unstd_posterior)
    
    # plot
    plt.scatter(x=p_grid, y=posterior)
    plt.plot(p_grid, posterior)
```

```python
plot_binomial_masked(3, 3)
plot_binomial_masked(3, 4)
plot_binomial_masked(5, 7)
```

### 2M4

Three double sided cards - BB, BW, WW. What is P(BB|B)? That is, if a card is selected at random, what is the probability that the other side of the card is black?

| Conjecture | Ways | Plausibility |
| --- | --- | ---|
| BB | 2 | 2/3 |
| BW | 1 | 1/3 |
| WW | 0 | 0/3 |

$\therefore P(BB|B) = 2/3$

### 2M7

Weighted cards such that the probability of a BW is twice that of a BB, and WW is three times that of a WW. What is P(BB|B) now?

| Conjecture | Counts | Plausibility |
| --- | --- | --- |
| BB, BW | (2)(1) = 2 | 2/8 |
| BB, WW | (2)(2) = 4 | 4/8 |
| BW, BB | (1)(0) = 0 | 0/8 |
| BW, WW | (1)(2) = 2 | 2/8 |
| WW, BB | (0)(0) = 0 | 0/8 |
| WW, BW | (0)(1) = 0 | 0/8 |

$P(BB, \_) = P(BB, BW) + P(BB, WW)$

$P(BB, \_) = 2/8 + 4/8$

$\therefore P(BB, \_) = 0.75$

### 2H1

* $P(A)=0.5$
* $P(B)=0.5$
* $P(twins|A)=0.1$
* $P(twins|B)=0.2$

So in this question we need to determine $P(2\_twins|1\_twins) = \frac{P(1\_twins, 2\_twins)}{P(1\_twins)}$

Denominator:

$P(1\_twins)=\frac{(0.1)(0.5) + (0.2)(0.5)}{2}=0.15$

Numerator:

$P(1\_twins, 2\_twins)=P(1\_twins, 2\_twins | A) + P(1\_twins, 2\_twins | B)$

$P(1\_twins, 2\_twins)=(0.1)(0.1)(0.5) + (0.2)(0.2)(0.5)$

$P(1\_twins, 2\_twins)=0.025$

Solved:

$P(2\_twins|1\_twins) = 0.025/0.15$

$\therefore P(2\_twins|1\_twins) = 0.167$
